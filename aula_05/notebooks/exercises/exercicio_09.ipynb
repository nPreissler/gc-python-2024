```python
import pandas as pd

df_filmes = pd.read_json('../../aula_03/exercício_07/Filmes.json')
df_series = pd.read_json('../../aula_03/exercício_07/Séries.json')
df_desenhos = pd.read_json('../../aula_03/exercício_07/Desenhos.json')

df_geral = pd.concat([df_filmes, df_series, df_desenhos]).reset_index(drop=True)
df_geral
```

```python
df_melhores = df_geral[df_geral['Nota'] == max(df_geral['Nota'])]
df_melhores
```
