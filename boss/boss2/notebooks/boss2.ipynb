```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pokedex = pd.read_csv('../pokedex_(Update_05.20).csv')
pokedex.dropna()
```

# Número de Pokémon por geração:

Crie uma visualização que mostre o número de Pokémon 
por geração, mas vá além de um gráfico simples de barras.

```python
gen_counts = pokedex['generation'].value_counts().sort_index()

plt.figure(figsize=(8, 5))
plt.scatter(
    x=gen_counts.index, 
    y=gen_counts.values, 
    s=gen_counts.values * 10,
    alpha=0.6, 
    c='blue'
)

for i, value in enumerate(gen_counts.values):
    plt.text(gen_counts.index[i], value + 0.1, str(value),
             ha='center', va='bottom', fontsize=9)

plt.title('Número de Pokémon por Geração')
plt.xlabel('Geração')
plt.ylabel('Número de Pokémon')
plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)
plt.tight_layout()
plt.show()
```
