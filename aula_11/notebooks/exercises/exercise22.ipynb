```python
import pandas as pd 
```

```python
from sklearn.preprocessing import StandardScaler
```

## Normalize o atributo Rendimentos utilizando Re-escala [0,1].


```python
data = {
    "Idade": [37, 46, 62, 31, 61, 50, 54, 27, 22, 56],
    "Casado": ["SIM", "NÃO", "SIM", "SIM", "SIM", "SIM", "NÃO", "NÃO", "NÃO", "SIM"],
    "Filhos": [2, 0, 0, 1, 3, 2, 0, 0, 0, 3],
    "Financiamento": ["NÃO", "SIM", "NÃO", "SIM", "SIM", "NÃO", "SIM", "NÃO", "NÃO", "SIM"],
    "Rendimentos": [17729.80, 41016.00, 26909.20, 22522.80, 57880.70, 16497.30, 38446.60, 15538.80, 12640.30, 41034.00],
    "Sexo": ["M", "F", "M", "M", "F", "F", "M", "M", "M", "F"]
}

df = pd.DataFrame(data)

df
```

```python
scaler = StandardScaler()
df['Rendimentos reescalados'] = scaler.fit_transform(df[['Rendimentos']])
df
```

## Encontra o menor e o maior valor do atributo e atribui os valores extremos da nova escala

```python
max_index = df['Rendimentos'].idxmax()
min_index = df['Rendimentos'].idxmin()

df.at[max_index, 'Rendimentos reescalados'] = 1
df.at[min_index, 'Rendimentos reescalados'] = 0
```
