# Exercício Guiado:Calcular medias de tendência central e dispersão com Pandas 

Realizar uma analise das medidas estatisticas no conjunto de dados de cidades brasileiras com mais de 100 mil habitantes. Deve primeiro remover as cidades com números extremos do conjunto usando a medida de Desvio Padrão, depois calcular a Media e Mediana de população nestas cidades, e a Moda de Unioes Federativas presentes.

```python
import pandas as pd
```

## Leitura dos Dados

```python
dados = pd.read_html('https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_do_Brasil_acima_de_cem_mil_habitantes')
```

```python
dados
```

```python
# Queremos apenas a primeira tabela da extração. Cuidado! Execute está célula apenas uma vez.
dados_populacao = dados[0]
dados_populacao
```

```python
dados_populacao.columns = ['ranking', 'municipio', 'uf', 'populacao']  
dados_populacao
```

```python
dados_populacao[:20]
```

```python
dados_populacao.dtypes
```

```python
dados_populacao['populacao'] = dados_populacao['populacao'].str.replace(r'\s', '', regex=True).astype(int)
dados_populacao
```

## Calculando o Desvio Padrão

```python
media_total = dados_populacao['populacao'].mean()
mediana_total = dados_populacao['populacao'].median()

print(f'Media conjunto de dados completo: {media_total:,}\nMediana conjunto de dados completo: {mediana_total:,}')
```

```python
desvio_padrao = dados_populacao['populacao'].std()

print(f'Desvio Padrão conjunto de dados completo: {desvio_padrao:,}')
```

## Filtrando Valores Extremos

```python
limite_superior = media_total + 2*desvio_padrao
limite_inferior = media_total - 2*desvio_padrao

print(f'Serão excluídos os valores fora da faixa [{limite_inferior:,}, {limite_superior:,}]')
```

```python
dados_populacao['dentro_da_faixa'] = dados_populacao['populacao'] <= limite_superior
dados_populacao
```

```python
dados_populacao_reduzido = dados_populacao[dados_populacao['dentro_da_faixa']]
dados_populacao_reduzido
```

## Calculando Medidas para o novo Conjunto

```python
media_reduzido = dados_populacao_reduzido['populacao'].mean()
mediana_reduzido = dados_populacao_reduzido['populacao'].median()

print(f'Media conjunto de dados reduzido: {media_reduzido:,}\nMediana conjunto de dados reduzido: {mediana_reduzido:,}')
```
