```python
import pandas as pd
```

```python
books_df = pd.read_csv('../../files-dataframes/livros.csv')
books_df

# carregue os arquivos livros.csv e emprestimos.csv em df's separados
```

```python
loans_df = pd.read_csv('../../files-dataframes/emprestimos.csv')
loans_df

# carregue os arquivos livros.csv e emprestimos.csv em df's separados
```

```python
# realizar um inner join entre as tabelas, unindo as pela coluna "id_livro"

merged_df = pd.merge(books_df, loans_df, on='id_livro', how='inner') 
merged_df
```
